#### 4.3.3.1 Suppression des produits avec des valeurs >100 dans au moins l'une des colonnes [*_100g]

On retire préalablement les colonnes qui peuvent légitimement contenir des valeurs supérieures à 100.

Les colonnes restantes sont enregistrées dans la variable '**newListCols100g**'.

```python
colsAExclure = ['energy-kcal_100g','energy_100g','nutrition-score-fr_100g']
newListCols100g = [ele for ele in listCols100g if ele not in colsAExclure] 
print(newListCols100g) # Affichage des colonnes qui ne doivent pas contenir de valeur supérieure à 100.
```

Je crée une fonction qui filtre le DF donnée en entrée en retirant tous <br>
les enregistrements qui contiennent au moins une valeur supérieure <br>
à 100 parmi les colonnes à contrôler.

```python
def filtreValeursAberantes(df, cols, value, inf=True):
  '''Prend un df en entrée et retourne un df filtré
Si inf vaut True alors seules les lignes qui ont toutes 
leurs valeurs inférieures ou égales à 'value', dans 
l'ensemble des colonnes 'cols', sont conservées
Sinon, seules les lignes qui ont toutes 
leurs valeurs supérieur ou égales à 'value', dans 
l'ensemble des colonnes 'cols', sont conservées'''
  #Initialisation de la liste d'index à retourner
  if inf :
    listeIndexValide = df[cols].fillna(0).apply(lambda row: row.name if (row<=value).all() else None, axis=1)
  else:
    listeIndexValide = df[cols].fillna(0).apply(lambda row: row.name if (row>=value).all() else None, axis=1)
  listeIndexValide.dropna(inplace=True)
  listeIndexValide = [int(i) for i in listeIndexValide]
  return df[df.index.isin(listeIndexValide)]
```

#### 4.3.4.4 Définition des fonctions de remplissage des données manquantes

```python
def calcMeanByCateg(df,cols,categ):
  '''Retourne un DF contenant les moyennes 
  calculées de chaque 'cols' classé selon les modalités de la colonne 'categ' 
  * df est le DataFrame 
  * cols corresponds à une liste de noms de colonnes
  * categ correspond au nom d'une colonne'''
  dictValRef = {}
  for cat in df[categ].unique():
    dictValRef.update({cat: df[cols][df[categ] == cat].mean()})
  return pd.DataFrame(dictValRef).T



```

```python
def fillMissingValues(df, cols, categ):
  '''Renseigne les valeurs manquantes des colonnes **cols** 
  du DataFrame 'df' à partir des moyennes de ces colonnes.
  Les valeurs sont renseignées par catégories selon les modalités de la colonne **categ**'''

  for categorie in df[categ].unique():
    df.loc[df[categ]==categorie, cols] = df[df[categ]==categorie][cols].fillna(df[df[categ] == categorie][cols].mean())
  return df
```

## 5.3 KNN Regressor

Le même processus est réalisé ici avec l'algorithme **KNN Regressor**.<br>
A la différence de l'algorithme **LinearRegression**, l'algorithme **KNN Regressor** <br>accepte en entrée plusieurs hyperparamètre qui vont influer sur le taux de réussite en sortie.<br>
Pour m'aider à faire le meilleur choix des hyperparamètres, j'utiliserai la méthode **GridSearchCV**.

Nous réutilisons le même DF '**dataRL**' utilisé précédemment ainsi <br>que la même méthodologie de création des données d'entrainement et de test.

```python
print(dataRL['nutrition-score-fr_100g'].notnull().sum()/len(dataRL))
dataRL.head()
```

```python
# Création des features
X = dataRL[listCols100g]
# Création des labels
y = dataRL['nutrition-score-fr_100g']
```

```python
# Création du train set et du test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(X_train.shape)
print(X_test.shape)
```

Je réalise un premier essai avec l'ensemble des valeurs des hyperparamètres par défaut :

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

J'obtiens un score d'un peu plus de **86%** de réussite sur les données de test.<br>
Le choix de l'algorithme **KNN Regressor**, même non optimisé, me donne un <br>résultat amélioré d'un peu plus de **24%** par rapport à l'algorithme **LineaRegression**.

Je cherche maintenant à optimiser mon algorithme en testant automatiquement, <br>grâce à la méthode GridSeachCV, la valeur des différents **hyperparamètres** <br>et en effectuant une **Cross-Validation**.

La **Cross-Validation** est un technique qui permet à notre algorithme <br>d'apprendre et de tester ses résultats sur différentes partie du jeux de donnée.<br>
Si lors de la phase d'apprentissage l'algorithme utilise **80%** du jeu de donnée <br>et se test sur les **20%** restant, la **Cross-Validation** permettra de répéter <br>le processus en alternant le choix des données utilisé dans le **train-test** <br>et le **test-set** jusqu'à ce que toutes les données ait été utilisé à la fois <br>dans le **train-test et le test-set**, éliminant ainsi les erreurs liées au choix, <br>aléatoire, des données à inclure dans le **train-set** et le **test-set**.

```python
# Test des différentes valeurs des hyperparamètres et Cross-Validation avec GridSearchCV
from sklearn.model_selection import GridSearchCV
# Dictionnaire contenant les différents hyperparamètres à régler 
# ainsi que chaque valeur à tester pour ces hyperparamètres.
param_grid = {'n_neighbors': np.arange(1,20),
              'metric': ['euclidean', 'manhattan'],
              'weights': ['uniform', 'distance']}
grid = GridSearchCV(KNeighborsRegressor(), param_grid, cv=5)

grid.fit(X_train, y_train)
```

On affiche maintenant les meilleurs valeurs des hyperparamètres testés :

```python
print('Meilleurs paramètres:',grid.best_params_)
print('Meilleur score:', grid.best_score_)
```

Sauvegarde du modèle avec les meilleurs paramètres :

```python
model = grid.best_estimator_
```

Test de notre modèle sur les données de test :

```python
model.score(X_test, y_test)
```

Notre modèle est maintenant optimisé et nous obtenons un taux de réussite <br>de plus de **88,6%** soit **2,6%** de mieux que notre modèle non optimisé <br>ou **26,6%** de mieux que l'algorithme **LinearRegression**.

### Learning Curve

Je cherche à savoir si la quantité de donnée contenu dans mon DataFrame <br>est suffisante et si la qualité de mon modèle pourrait être amélioré par <br>l'ajout de nouvelles données.

Pour répondre à cette question j'utilise la **Learning Curve** qui me permettra <br>de visualiser la **performance de mon modèle** en fonction de la quantité <br>de données contenu dans mon **validation-set**.

L'apport de nouvelles données peut-être une opération couteuse en temps <br>et/ou financièrement. Il est important de savoir s'il est nécessaire d'ajouter <br>de nouvelles données ou si au contraire cela constitue une perte de temps et/ou d'argent.

```python
from sklearn.model_selection import learning_curve
N, train_score, val_score = learning_curve(model, X_train, y_train, 
                                           train_sizes = np.linspace(0.1, 1.0, 10), cv=5)
plt.plot(N, train_score.mean(axis=1), label='train')
plt.plot(N, val_score.mean(axis=1), label='validation')
plt.xlabel('train_sizes')
plt.legend()
```

Comme nous pouvons l'observer, la courbe d'apprentissage <br>n'a pas encore atteint son point d'aplatissement et les résultats <br>de l'algorithme pourrait être amélioré par l'apport de nouvelles données.

## 5.4 KNN Classifier

J'utilise maintenant l'algorithme **KNeighborsClassifier** pour compléter <br>la colonne '**nutriscore_grade**'.<br>
La colonne '**nutriscore_grade**' contient des valeurs discrètes, pour cette <br>raison j'utilise un algorithme de classification.

<u>Le processus est pratiquement le même que pour l'utilisation de l'algorithme **KNeighborsRegressor**</u> :
1. *Nouveau* : Notre algorithme fonctionne avec des valeurs numériques uniquement. <br>Je commence par remplacer les lettres '**a**' à '**e**' contenu dans la colonne <br>'**nutriscore_grade**' par les valeurs numériques correspondantes allant de '**1**' à '**5**'
2. Sélection des features (X) et des labels (y)
3. Séparation du jeu de donnée en train-set / test-set
4. Test de l'algorithme avec les valeurs d'hyperparamètres par défaut

9. Affichage de la learning Curve

### 5.4.1 Sélection des features et des labels et <br>remplacement des valeur de 'nutriscore_grade' <br>par des valeurs numériques

Création d'un DataFrame sans valeur null pour la colonne ''nutriscore_grade'

```python
print(data['nutriscore_grade'].notnull().sum()/len(data))
dataML = data[data['nutriscore_grade'].notnull()]
print(dataML['nutriscore_grade'].notnull().sum()/len(dataML))
```

J'effectue le remplacement des valeurs de '**nutriscore_grade**' directement dans le DF contenant les labels.

```python
# Sélection des features
X = dataML[listCols100g]
# Sélection des labels
y = dataML['nutriscore_grade'].replace(['a','b','c','d','e'],[1,2,3,4,5])
```

### 5.4.2 Séparation du jeu de donnée en train-set / test-set

```python
# Création du train set et du test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

### 5.4.4 Cross-Validation et recherche des meilleurs réglages des hyperparamètres avec GridSearchCV

```python
# Test des différentes valeurs des hyperparamètres et Cross-Validation avec GridSearchCV
from sklearn.model_selection import GridSearchCV
# Dictionnaire contenant les différents hyperparamètres à régler 
# ainsi que chaque valeur à tester pour ces hyperparamètres.
param_grid = {'n_neighbors': np.arange(1,20),
              'metric': ['euclidean', 'manhattan'],
              'weights': ['uniform', 'distance']}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)

grid.fit(X_train, y_train)
```

```python
print('Meilleurs paramètres:',grid.best_params_)
print('Meilleur score:', grid.best_score_)
```

### 5.4.5 Enregistrement et tests des meilleurs paramètres dans notre modèle

```python
# Sauvegarde du modèle avec les meilleurs paramètres
model = grid.best_estimator_
```

```python
# Test de notre modèle sur les données de test
model.score(X_test, y_test)
```

Le modèle optimisé à un taux de réussite de plus de 78.1%

### 5.4.8 Renseignement des valeurs manquantes pour nutriscore_grade

Je crée ici une fonction qui appel la méthode '**predict**' de notre modèle <br>puis retourne la lettre correspondant au chiffre retourné par la méthode '**predict**' via l'utilisation d'un dictionnaire.

```python
def predict_nutriscore_grade(model, features):
  label_predict = model.predict(features)
  mapping = {1:'a', 2:'b', 3:'c', 4:'d', 5:'e'}
  return mapping[label_predict.item()]
```

```python
# Remplissage de la colonne 'nutriscore_grade'
data['nutriscore_grade'] = data.apply(lambda row: 
            predict_nutriscore_grade(model, np.array(row[listCols100g]).reshape(1,10)) \
            if pd.isnull(row['nutriscore_grade']) \
            else row['nutriscore_grade'] ,axis=1)
```
